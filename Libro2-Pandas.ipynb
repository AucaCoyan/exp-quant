# Filtrado (P. 37)
Selección de un rango de fechas

```python
data = pd.read_excel('UCEMA/AAPL.xlsx')

# Para setear el índice
data.set_index('timestamp', inplace=True)

# Para ordenar la tabla de viejo a nuevo
data = data[::-1]

data.loc["2000-03-26":"2000-04-02"]
```

Setear por filtros condicionales

```python
filtro = data.loc[data['volume']>50000000]
filtro
```

```python
# si le saco el loc, me devuelve un True / False en cada fila
filtro2 = data['volume']>50000000
filtro2
```

Para hacer un filtro complejo tengo que ir concatenando filtros

```python
filtro = data[data['volume']>50000000]
filtro = filtro[filtro['close']<100]
filtro
```

```python
# o tambien de otra forma
filtro = data[data['volume']>50000000].loc[data['close']<100]
filtro
```

```python
# O se puede definir 2 filtros y luego mergearlos (L)

filtro1 = data[data['volume']>50000000]

filtro2 = data[data['close']<100]

# merge
#   inner es la interseccion de ambos
#   outer es la union de ambos

filtro = filtro1.merge(filtro2, how="inner")
filtro
```

```python
# o  sino también usando & para AND y | para OR

filtro = data[(data['volume']>50000000) & (data['close']<100)]
filtro
```

# Funciones básicas con pandas

```python
import pandas as pd

data = pd.read_excel('UCEMA/AAPL.xlsx')

# Tengo que dar vuelta la tabla para calcular el % con respecto al día de ayer, sino sería al dia de mañana
data = data[::-1]
data['variacion'] = round(data['adjusted_close'].pct_change() *100 , 2)
data
```

```python
# le podemos pasar un argumento con la cantidad de días que querés que promedie hacia atrás
data['variacion'] = round(data['adjusted_close'].pct_change(3) *100 , 2)
data
```

```python
# o sea que si le ponés -1 te calcula con el dia de ayer, en una tabla sin ordenar

data = pd.read_excel('UCEMA/AAPL.xlsx')
data['variacion'] = round(data['adjusted_close'].pct_change(-1)*100, 2)
data
```

Con la funcion diff() calcula la diferencia numerica entre dos valores
Ojota que hay que usarlo sobre los cierres normales, no los ajustados, porque los ajustados tienen en cuenta balances splits y dividendos que no es lo real

```python
data['variacion_precio'] = data['close'].diff(-1)
data
```

Eval() te devueve True o False segun la condicion que le des.
Lo más prolijo en un sistema de trading es poner condiciones en columnas y que diga que condiciones pasaron y cuales no

```python
data['vela_verde'] = data.eval('close >= open')
# o por ejemplo poner una condicion de 3%
data['vela_verde'] = data.eval('close >= open * 1.03 or close <= open * 0.97')
data.head(8)
```

La función shift(N) nos devuelve el valor de N filas anteriores

# Para contar datos

```python
data = pd.read_excel('UCEMA\AAPL.xlsx')
data['gap_positivo'] = data.open > data.close.shift(-1)

positivos = 0
negativos = 0

for dato in data['gap_positivo']:
    if dato == True:
        positivos += 1
    else:
        negativos += 1

print(F"Positivos: {positivos}, Negativos: {negativos}")
```

Se puede solucionar de forma mas rapido, sin pasar fila por fila

```python
positivos = data.gap_positivo[data.gap_positivo == True].count()
negativos = data.gap_positivo[data.gap_positivo == False].count()
print(F"Positivos: {positivos}, Negativos: {negativos}")
```
