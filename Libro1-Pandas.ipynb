Para hacer un filtro complejo tengo que ir concatenando filtros

```python
filtro = data[data['volume']>50000000]
filtro = filtro[filtro['close']<100]
filtro
```

```python
# o tambien de otra forma
filtro = data[data['volume']>50000000].loc[data['close']<100]
filtro
```

```python
# O se puede definir 2 filtros y luego mergearlos (L)

filtro1 = data[data['volume']>50000000]

filtro2 = data[data['close']<100]

# merge
#   inner es la interseccion de ambos
#   outer es la union de ambos

filtro = filtro1.merge(filtro2, how="inner")
filtro
```

```python
# o  sino también usando & para AND y | para OR

filtro = data[(data['volume']>50000000) & (data['close']<100)]
filtro
```

# Para contar datos

```python
data = pd.read_excel('UCEMA\AAPL.xlsx')
# al parecer podés acceder a las columnas de pandas con . como si fueran objetos
data['gap_positivo'] = data.open > data.close.shift(-1)

positivos = 0
negativos = 0

for dato in data['gap_positivo']:
    if dato == True:
        positivos += 1
    else:
        negativos += 1

print(F"Positivos: {positivos}, Negativos: {negativos}")
```

Se puede solucionar de forma mas rapido, sin pasar fila por fila

```python
positivos = data.gap_positivo[data.gap_positivo == True].count()
negativos = data.gap_positivo[data.gap_positivo == False].count()
print(F"Positivos: {positivos}, Negativos: {negativos}")
```

Desviación estandar
Vamos a calcular para cada fecha de la tabla el desvío estandar o volatilidad de los rendimientos diarios, y luego vamos a anualizar multiplicándolo por la raíz de la cantidad de ruedas que hay en un año.

```python
import math
data = pd.read_excel('UCEMA/AAPL.xlsx')
data = data.sort_values(by='timestamp', ascending=True)

res = pd.DataFrame()
res['fecha'] = data['timestamp']
res['variacion'] = round(
    data['adjusted_close'].pct_change()
                        * 100, 2)
res['sigma_40'] = round(res.variacion.rolling(40).std() * math.sqrt(250), 2) # Dividido 250 porque es la cantidad de ruedas que hay en un año
res = res.dropna()
res
```

Una vez que tenemos el sigma puntual de cada fecha para sus ultimas 40 ruedas vamos a calcular la mm mensual de esa volatilidad (media movil mensaual nos referimos a la volatiliad de 40 ruedas promedio de las últimas 20). La tabla va a arrancar un mes más tarde porque necesita tomar 20 datos para calcular el primer sigma.

```python
res['sigma_40_rolling'] = round(res.sigma_40.rolling(20).mean(), 2)
res = res.dropna()

res
```

# Funciones acumulativas
`cummax()` el máximo acumulado (por ejemplo máximo historico)

`cummin()` el minimo

`cumsum()` suma acumulada (armado de subtotales por fecha)

`cumprod()`el producto acumulado (ideal para rendimiento compuesto)

```python
data['maxHist'] = data.adjusted_close.cummax()
data['minHist'] = data.adjusted_close.cummin()
data
```

```python
data['volumenAcum'] = data.volume.cumsum()/1000000
data.drop(['maxHist', 'minHist'], axis=1, inplace=True)
data.round(2).dropna()
data
```

Rendimiento por año

```python
import datetime
import calendar

# data = misfunciones.leerExcel('UCEMA/AAPL.xlsx')
data = pd.read_excel('UCEMA/AAPL.xlsx')
data = data[::-1]

años = []
for i in range(2005, 2016):
    años.append(i)

rendimientos = pd.DataFrame(index=años)
for año in años:
    desde = pd.Timestamp(año - 1, 12, 31)
    hasta = pd.Timestamp(año + 1, 1, 1)

    if desde.weekday() == 6:
        desde = pd.Timestamp(año - 1, 12, 29)
    elif desde.weekday() == 5:
        desde = pd.Timestamp(año - 1, 12, 30)
    
    res = data.filter(['timestamp', 'adjusted_close'], axis=1 )
    res = data.set_index('timestamp').dropna()
    resFiltro = res.loc[ (res.index >= desde) & (res.index <= hasta) ].copy()
    resFiltro['factor'] = resFiltro.adjusted_close/resFiltro.adjusted_close.shift(1)
    resFiltro['Yield'] = (resFiltro['factor'].cumprod() - 1) * 100
    rend = float(resFiltro[-1:]['Yield'])
    rendimientos.loc[años, 'Yield'] = round(rend, 2)

rendimientos
```
