Crear una columna a partir de otras columnas.
(Uso la base de datos de AAPL.xlsx ahora)

```python
data = pd.read_excel('UCEMA/AAPL.xlsx', sheet_name='Hoja1')
data['precio_medio'] = (data.open + data.close + data.low + data.high) / 4

data['vol_mln_usd'] = round(data.volume * data.precio_medio) / 1000000
data
```

Creacion de una columna con datos del df

```python
for i in range(len(data)):
    if data.loc[i,'close'] >= data.loc[i, 'open']:
        data.loc[i,'color_vela'] = "verde"
    else:
        data.loc[i, 'color_vela'] = "rojo"

data
```
