# Ejercicios de la página 126

```python
import csv
data = csv.reader(open('UCEMA/balances.csv'),delimiter=';')
balances = [fila for fila in data]

empresas = ["AAPL", "AMZN", "FB", "TSLA", "KO", "NFLX"]
screener = {}

for empresa in empresas:
    for balance in balances:
        if balance[0]==empresa and balance[1]=="anual":
            try:
                screener[empresa] = round(int(balance[16]) / int(balance[13]),2)
            except:
                screener[empresa] = "No se pudo calcular"
            break

print(screener)
for linea in balances:
    print(balances[linea])
```

```python
reader = csv.reader(open('UCEMA/SPY.csv'), delimiter=';')
data = list(reader)
print(data[0][5])

# preparo la lista
ad_close = []
data = data[1:]
for row in data:
    ad_close.append(float(row[5]))

# print(ad_close)

base = ad_close[0]
counter = 0
for price in ad_close:
    print(price)
    print("el precio base es", base)
    print("el precio a superar es", base*1.05)
    if price >= base*1.05:
        counter = counter + 1
        base = price
    else:
        base = price

print(counter)
    
```
